# Offense Data Analytics

Import python pandas

```python
import pandas as pd
```

Read offense data file.

```python
df2 = pd.read_excel("Offenses Data Test.xlsx", header=0)
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

Rename dateReport (date) column to readable header of the dataframe.

```python
df2.rename(columns = {'dateReport (date)':'crime_date'}, inplace = True)
   
```

```python
df2.head()
```

Find month from the crime date.

```python
df2['month'] = df2['crime_date'].dt.month
```

```python
df2.head(50)
```

Remove duplicate rows but keep the last matching row.

```python
df3 = df2.drop_duplicates(keep='last')
```

```python
df2.shape
```

```python
df3.shape
```

```python
df3.head(50)
```

```python
pd.value_counts(df2['Category'])[:10]
```

Find the count of crime per crime category.

```python
crime_rate = df3.groupby('Category').size().sort_values(ascending=False).rename('counts').reset_index()
```

```python
crime_rate.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

f, ax = plt.subplots(figsize=(6, 15))


# Plot the total crimes
sns.set_color_codes("pastel")
sns.barplot(x="counts", y="Category", data=crime_rate,
            label="Total", color="b")

ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="Type",
       xlabel="Crimes")
sns.despine(left=True, bottom=True)

plt.show()
```

```python
ax = sns.countplot(x="month", data=df3,
                   facecolor=(0, 0, 0, 0),
                   linewidth=5,
                   edgecolor=sns.color_palette("dark", 24))
```

```python
crime_by_zone = df3.groupby('Zone').size().sort_values(ascending=False).rename('counts').reset_index()
```

```python
crime_by_zone.head(20)
```

```python
sns.set_color_codes("pastel")
sns.barplot(x="counts", y="Zone", data=crime_by_zone,
            label="Total", color="b")

ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="Type",
       xlabel="Zone")
sns.despine(left=True, bottom=True)

plt.show()
```
